```python
import pandas as pd
import openpyxl
import re

# Load Excel file
file_path = r"C:\Users\sowmi\OneDrive\Desktop\TamilNadu_Coconut_Varieties.xlsx"
df = pd.read_excel(file_path)

# Trim spaces in all string columns
for col in df.select_dtypes(include='object'):
    df[col] = df[col].map(lambda x: x.strip() if isinstance(x, str) else x)


# Fix Best_Season formatting
def fix_season_format(season):
    if isinstance(season, str):
        season = re.sub(r'([a-zA-Z])([A-Z])', r'\1, \2', season)
        return season
    return season

df["Best_Season"] = df["Best_Season"].apply(fix_season_format)

# Convert water columns to numeric
df["Basin irrigation water requirement(in liters)"] = pd.to_numeric(
    df["Basin irrigation water requirement(in liters)"], errors='coerce')
df["Drip Irrigation water requirement(in liters)"] = pd.to_numeric(
    df["Drip Irrigation water requirement(in liters)"], errors='coerce')

# Fill missing values
df.fillna("N/A", inplace=True)

# Export cleaned file (optional)
df.to_csv("Fully_Cleaned_Coconut_Varieties.csv", index=False,encoding='utf-8-sig')

# Preview cleaned data
df.head(37)
```
